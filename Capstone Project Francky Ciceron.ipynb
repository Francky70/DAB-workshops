```python
# Convert "Cohort Year" column to datetime formart
def convert_cohort(value):
    if pd.isna(value):
        return np.nan
    value = value.strip()
    if value.startswith('Fall'):
        year = value.split()[1]
        return pd.to_datetime(f'{year}-09-01')
    elif '-' in value:
        year = value.split('-')[0]
        return pd.to_datetime(f'{year}-09-01')
    else:
        return np.nan

df['Cohort Year Clean'] = df['Cohort Year'].apply(convert_cohort)
```

```python
# Merge PA County Table with the Cleaned Dataset
df_clean = pd.read_csv('capstone_cleaned.csv')
geo_table = pd.read_csv('PA_Zip_County.csv')
```

```python
# Trim the Zip code in the cleaned dataset to keep only 5 digits
df_clean['Zip Code'] = df_clean['Zip Code'].astype(str).str[:5]
```

```python
# Ensure Zip codes in County table are string
geo_table['ZIP Code'] = geo_table['ZIP Code'].astype(str)
```

```python
#Join both table on Zip Code
df_merged = df_clean.merge(geo_table[['ZIP Code', 'County']], left_on='Zip Code', right_on='ZIP Code', how='left')
df_merged['County'] = df_merged['County'].astype(str) + ' County'
```

```python
#Check for the new columns
df_merged.columns
```

```python
# Drop the Zip Code column from the County Table
df_merged.drop(columns=['ZIP Code'], inplace=True)
```

```python
# Check the first rows
df_merged.head()
```

```python
#Remove duplicate county at the end of the county name
df_merged['County'] = df_merged['County'].str.replace(r'\sCounty$', '', regex=True)
df_merged['County'] = df_merged['County'] + ' County'
```

```python
#Check the last 5 rows
df_merged.tail()
```

```python
# Check for duplicate Zip code
geo_table['ZIP Code'].value_counts().head(10)
```

```python
# Drop duplicate Zip Code
geo_table = geo_table.drop_duplicates(subset='ZIP Code')
```

```python
# Merge bothe table on Zip Code
df_merged = df_clean.merge(geo_table[['ZIP Code', 'County']], left_on='Zip Code', right_on='ZIP Code', how='left')
```
